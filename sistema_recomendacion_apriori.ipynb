```python
# Celda6 Generar recomendaciones para productos específicos
if sistema.rules is not None and len(sistema.rules) > 0:
    # Usar productos que realmente existen en el dataset
    productos_populares = sistema.df['product'].value_counts().head(3).index.tolist()
    print(f"Productos seleccionados para prueba: {productos_populares}")
    
    for producto in productos_populares:
        print(f"\n>>> Recomendaciones para: {producto}")
        sistema.generar_recomendaciones(producto, 3)
        print("-" * 50)
else:
    print("No hay reglas disponibles para generar recomendaciones")
```

```python
# Celda 7Experimentar con diferentes parámetros del algoritmo Apriori
print("EXPERIMENTACIÓN CON PARÁMETROS")
print("=" * 35)

print("1. Probando con parámetros más estrictos (menos reglas):")
sistema.ejecutar_apriori(min_support=0.04, min_confidence=0.6)
if sistema.rules is not None:
    print(f"Reglas generadas: {len(sistema.rules)}")
    sistema.mostrar_reglas_top(3)
else:
    print("No se generaron reglas con estos parámetros")

print("\n2. Probando con parámetros más flexibles (más reglas):")
sistema.ejecutar_apriori(min_support=0.02, min_confidence=0.3)
if sistema.rules is not None:
    print(f"Reglas generadas: {len(sistema.rules)}")
    sistema.mostrar_reglas_top(5)
else:
    print("No se generaron reglas con estos parámetros")
```

```python
# celda 8Análisis de métricas y resultados finales
print("ANÁLISIS DE MÉTRICAS Y RESULTADOS FINALES")
print("=" * 45)

if sistema.rules is not None and len(sistema.rules) > 0:
    # Gráfica de distribución de confianza
    plt.figure(figsize=(10, 6))
    plt.hist(sistema.rules['confidence'], bins=15, alpha=0.7, color='blue', edgecolor='black')
    plt.xlabel('Confianza')
    plt.ylabel('Frecuencia')
    plt.title('Distribución de Confianza en las Reglas')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    # Gráfica de distribución de lift
    plt.figure(figsize=(10, 6))
    plt.hist(sistema.rules['lift'], bins=15, alpha=0.7, color='green', edgecolor='black')
    plt.xlabel('Lift')
    plt.ylabel('Frecuencia')
    plt.title('Distribución de Lift en las Reglas')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    # Resumen estadístico
    print("RESUMEN ESTADÍSTICO DE LAS REGLAS:")
    print(f"- Total de reglas: {len(sistema.rules)}")
    print(f"- Confianza máxima: {sistema.rules['confidence'].max():.3f}")
    print(f"- Confianza mínima: {sistema.rules['confidence'].min():.3f}")
    print(f"- Lift máximo: {sistema.rules['lift'].max():.3f}")
    print(f"- Lift mínimo: {sistema.rules['lift'].min():.3f}")
    print(f"- Soporte máximo: {sistema.rules['support'].max():.3f}")
    print(f"- Soporte mínimo: {sistema.rules['support'].min():.3f}")
    
    # Reglas más fuertes
    print("\nREGLAS MÁS FUERTES (alto lift y confianza):")
    reglas_fuertes = sistema.rules[(sistema.rules['lift'] > 1.5) & (sistema.rules['confidence'] > 0.6)]
    if len(reglas_fuertes) > 0:
        for i, (idx, regla) in enumerate(reglas_fuertes.head().iterrows(), 1):
            antecedentes = list(regla['antecedents'])
            consecuentes = list(regla['consequents'])
            print(f"{i}. {antecedentes} → {consecuentes}")
            print(f"   Confianza: {regla['confidence']:.3f}, Lift: {regla['lift']:.3f}")
    else:
        print("No hay reglas que cumplan los criterios de fuerza")
        
else:
    print("No hay reglas disponibles para análisis")
```
